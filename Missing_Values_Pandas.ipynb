# Handling Missing Values in Python using Pandas

```python
# importing pandas and loading the data from excel file into dataframe df
```

```python
import pandas as pd
import numpy as np
df = pd.read_excel('BASEBALL_Manual.xlsx')
```

```python
# Checking the first 30 records from the data
```

```python
df.head(30)
```

```python
# 123Salary column contains some missing values shown by NaN.
# The goal is to replace the Nan with the mean of this column
```

```python
#Showing only the 123Salary Column using dataframe
df['123Salary']
```

```python
# Finding the mean of the 123Salary columne and placing it in a variable mean_value
mean_value = df['123Salary'].mean()
mean_value
```

```python
#Replacing the Nan in the original df by mean_value
df['123Salary'] = df['123Salary'].replace(np.nan, mean_value)
```

```python
# Checking first 30 records to see if the value has been replaced
df.head(30)
```
